**Method definition**

Def of methods for opening and creating the htsdata List of Dicts with no empty "htsno" keys

```python
from typing import List, Dict, Any
import json, re

def readHtsDataFile(path: str) -> List[Dict[str, Any]]:

    with open(path, 'r') as file:
       return json.loads(file.read())

def removeEmptyKeys(htsdata: List[Dict[str, Any]]) -> List[Dict[str, Any]]:
    

    result = []

    for indx, data in enumerate(htsdata):
        
        filtered_data = {key: value for key, value in data.items() if bool(value)}
        result.append(filtered_data)

    return result

def saveProcessedFile(htsdata: List[Dict[str, Any]], path: str):

    with open(path, 'w') as file:
        file.write(json.dumps(htsdata, indent=4))
```

```python
htsdata = removeEmptyKeys(readHtsDataFile('../../db_hts/htsdata/htsdata.json'))
saveProcessedFile(htsdata, 'processed_hts.json')
htsdata
```
