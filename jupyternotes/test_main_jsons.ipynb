```python
import json
import os

def gatherJSON(path: str):

    with open(path, 'r') as file:

        record = json.loads(file.read())

    return record
```

```python
filenames = sorted(os.listdir('test_files/'))

filenames
```

```python
final_list = []
list_of_empty = []

for file_name in filenames:
    data = gatherJSON(f'test_files/{file_name}')

    empty = 0

    for record in data:
        if(record['htsno'] == ''):
            empty += 1
    
    list_of_empty.append((file_name, empty))

    final_list.append(data)
```
