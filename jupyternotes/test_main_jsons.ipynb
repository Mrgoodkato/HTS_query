This is a notebook dedicated to testing the htsdata.json or hts parsed records in json format

```python
import json
import os

def gatherJSON(path: str):

    with open(path, 'r') as file:

        record = json.loads(file.read())

    return record
```

```python
filenames = sorted(os.listdir('test_files/'))

filenames
```

```python
final_list = []
list_of_empty = []

for file_name in filenames:
    data = gatherJSON(f'test_files/{file_name}')

    empty = 0
    counter = 0
    indent_first_record = False

    for record in data:
        if(record['htsno'] == ''):
            empty += 1
        
        if counter == 0 and record['indent'] == 0:
            indent_first_record = True
        
        if counter > 0 and record['indent'] == 0:
            indent_first_record = False
        
        counter += 1

        
    
    list_of_empty.append((file_name, empty, indent_first_record))

    final_list.append(data)
```
