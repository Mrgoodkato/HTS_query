```python
import json

def openJSON(path: str) -> list[dict, any]:
    """Function that opens a JSON file and returns that file as a list

    Args:
        path (str): Path to the JSON file

    Returns:
        list: Returns a list object with the JSON information
    """

    with open(path, 'r') as file:
        result = json.loads(file.read())
    
    return result
```

```python
empty_hts = openJSON('test_file.json')

empty_hts
```

```python
for key in empty_hts.keys():

    for index_records, record in enumerate(empty_hts[key]):
        
        prev_hts = record['prevhts']['indent']
        current_hts = record['nohts_rec']['indent']
        next_hts = record['nexthts']['indent']


        if prev_hts == current_hts and current_hts == next_hts:
            empty_hts[key][index_records]['type'] = 'EQUAL-EQUAL'
        elif prev_hts < current_hts and current_hts > next_hts:
            empty_hts[key][index_records]['type'] = 'LESS-MORE'
        elif prev_hts > current_hts and current_hts < next_hts:
            empty_hts[key][index_records]['type'] = 'MORE-LESS'
        elif prev_hts > current_hts and current_hts == next_hts:
            empty_hts[key][index_records]['type'] = 'MORE-EQUAL'
        elif prev_hts < current_hts and current_hts == next_hts:
            empty_hts[key][index_records]['type'] = 'LESS-EQUAL'
        elif prev_hts == current_hts and current_hts < next_hts:
            empty_hts[key][index_records]['type'] = 'EQUAL-LESS'
        elif prev_hts == current_hts and current_hts > next_hts:
            empty_hts[key][index_records]['type'] = 'EQUAL-MORE'

print(empty_hts['0101.json'])
```
